Importing Libraries : Pandas and Numpy

```python
import pandas as pd
import numpy as np
```

Reading the csv file

```python
df = pd.read_csv('/content/Sample - Superstore.csv', encoding='latin1')
print(df.head())
```

Printing shape of dataset (printing number of rows and colomn)

```python
print("Shape of dataset:", df.shape)
```

Checking Empty values

```python
print(df.isnull().sum())
```

Checking Duplicate Values

```python
print(df.duplicated().sum())
```

```python
df = df.drop_duplicates()
print(df.duplicated().sum())
```

No duplicate Values

```python
print(df.dtypes)
```

Converting the dates

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
print(df[['Order Date', 'Ship Date']])
```

Order by dates

```python
df = df.sort_values(by='Order Date')
print(df[['Order Date', 'Ship Date']])
```

Resetting index

```python
df = df.reset_index(drop = True)
print(df.head())
```

Saving cleaned csv file

```python
df.to_csv('cleaned_superstore.csv', index=False)
print("Data cleaning completed successfully. File saved as 'cleaned_superstore.csv'")
```
